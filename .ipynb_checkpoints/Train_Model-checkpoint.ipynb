## <font color="blue"> Data Modeling</font>

### <font color="green">Separation des caractérisques et de la cible</font>

```python
X = data_aqi_new.drop(columns=['Air Quality'])
y = data_aqi_new['Air Quality']
```

### <font color="green">Labelisation de la cible</font>

```python
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

### <font color="green">Données d'entrainement et de test</font>

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=40)
print(f"X_train :{X_train.shape}")
print(f"y_train :{y_train.shape}")
print(f"X_test :{X_test.shape}")
print(f"y_test :{y_test.shape}")
```

### <font color="green">Entrainement du modèle</font>

```python
models = {
    "Logistic Regression": LogisticRegression(solver='saga', max_iter=1000),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=40),
    "Support Vector Machine": SVC(),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
}

results = {}
```

#### <font color="blue">Performance des modèles</font>

```python
for name, model in models.items():
    model.fit(X_train, y_train)  
    y_pred = model.predict(X_test) 
    joblib.dump(model, f'{name}.joblib')
    acc = accuracy_score(y_test, y_pred)
    pres = precision_score(y_test, y_pred, average='weighted')
    rec = recall_score(y_test, y_pred, average='weighted')
    results[name] = acc  
    print(f"\n🏆 {name}\n")
    print(f"          - Accuracy: {acc:.4f}")
    print(f"          - Pression score: {pres:.4f}")
    print(f"          - Recall score: {rec:.4f}\n")
    print(classification_report(y_test, y_pred))
```

#### <font color="blue">Visualisation des modèles</font>

```python
results_df = pd.DataFrame(results.items(), columns=["Model", "Accuracy"])
plt.figure(figsize=(8, 5))
sns.barplot(x="Accuracy", y="Model", data=results_df, palette="viridis")
plt.xlabel("Accuracy Score")
plt.ylabel("Model")
plt.title("Compare Accuracy Score of 4 Model")
plt.xlim(0, 1)
plt.show()
```

### <font color="green">Test manuel de meilleur paramètre</font>

```python
model_test = SVC(kernel='poly', C=10, gamma=100)
model_test.fit(X_train, y_train)
y_pred_test = model_test.predict(X_test)
```

```python
accuracy_score(y_test, y_pred_test)
```

### <font color="green">Hyper Paramètre du Meilleure modèle : Support Vector Machine(SVM)</font>

```python
degreeG = [0, 1, 6]
cSetG = [0.1, 1, 10]
gammaG = [0.1, 1, 10]
kernelsG = ['linear', 'rbf', 'poly']


from sklearn.model_selection import GridSearchCV
model = SVC()
gridSearch = GridSearchCV(estimator=model,param_grid={'degree':degreeG,'kernel':kernelsG,'C':cSetG,'gamma':gammaG},cv=4)
gridSearch.fit(X_train, y_train)
print(f'The best parameter is {gridSearch.best_params_}')
print(f'The best score is {gridSearch.best_score_}')
print(f'The best estimator is {gridSearch.best_estimator_}')
```
